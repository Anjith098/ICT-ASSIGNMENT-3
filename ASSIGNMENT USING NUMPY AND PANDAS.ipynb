```python
#QUESTION 3 - PANDAS 

print("Missing Value : ")
df.isnull().sum()
```

```python
hp = df[(df['horsepower']!= '?')]['horsepower'].astype(int).median()
hps = df['horsepower'].values
hps[hps=='?'] = hp
df['horsepower']= hps.astype(int)
df['horsepower']
```

```python
#QUESTION 4- pandas
df["power_to_weight_ratio"]=df["horsepower"]/df["weight"]

```

```python
#QUESTION 5- PANDAS

grouped_cars = df.groupby('origin')['mpg'].mean()
grouped_cars
```

```python
#QUESTION 6- PANDAS 

top_10_mpg = df.sort_values('mpg', ascending=False).head(10)
top_10_mpg
def performance_score(row):
    return row['mpg'] * row['acceleration'] / row['weight']
df['performance_score'] = df.apply(performance_score, axis=1)
df.head()
```

```python
#QUESTION 8 - PANDAS 

summary_df = df.groupby('model year').agg({'mpg': 'mean', 'weight': 'mean', 'horsepower': 'mean'})
summary_df
```

```python
#QUESTION 9- PANDAS 


filtered_df = df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]
filtered_df.to_csv('high_mpg_cars.csv', index=False)
print("Filtered data saved to high_mpg_cars.csv")
```

```python
#QUESTION - 10 -PANDAS

Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['mpg'] < lower_bound) | (df['mpg'] > upper_bound)]
outlier_df = outliers[['car name', 'mpg', 'model year']]
outlier_df
```
